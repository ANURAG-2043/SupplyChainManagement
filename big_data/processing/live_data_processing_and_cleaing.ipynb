```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
import glob
```

```python
# Define the folder path and pattern to match CSV files
file_pattern = 'C:/Users/kat92/Documents/GitHub/SupplyChainManagement/data/raw/*.csv'

# Use glob to find all CSV files matching the pattern
file_paths = glob.glob(file_pattern)
```

```python

column_names = [
    'Product Type', 'SKU', 'Price', 'Availability', 'Number of products sold',
    'Revenue generated', 'Customer demographics', 'Stock levels', 'Lead times',
    'Order quantities', 'Shipping times', 'Shipping carriers', 'Shipping costs',
    'Supplier name', 'Location', 'Production volumes', 'Manufacturing lead time',
    'Manufacturing costs', 'Inspection results', 'Defect rates', 'Transportation modes',
    'Routes', 'Costs', 'Timestamp'
]
```

```python
base_filename='processed_data'
output_dir=r'C:\Users\kat92\Documents\GitHub\SupplyChainManagement\data\processed'
i = 1
    
 #Loop through each file path and process it
for file_path in file_paths:
    
    # Read the CSV file
    data = pd.read_csv(file_path, header=None)
    data.columns = column_names
    
    # Define the filename with increasing index
    filename = f"{base_filename}_{i}.csv"
    filepath = os.path.join(output_dir, filename)
    
    # Save to CSV
    data.to_csv(filepath, index=False)
    print(f'Saved {filepath}')
    
    # Increment the counter
    i += 1
```
